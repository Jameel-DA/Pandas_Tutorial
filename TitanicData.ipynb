Data Transformation

```python
# How to upper case column

df2['sex'] = df2.sex.apply(func=str.upper)
df2['embark_town'] = df2.embark_town.apply(func=str.upper)
df2
```

```python
# Create Custom Column based on Existing Column

def age_classification(age):
    if age < 18:
        return "child"
    elif age < 50:
        return 'adult'
    else:
        return 'senior citizen'
    
df2['age Category'] = df2.age.apply(func=age_classification)  
df2.head(60)   


def fare_classification(fare):
    if fare < 50:
        return '0-50'
    elif fare < 100:
        return '50-100'
    elif fare < 200:
        return '100-200'
    elif fare < 300:
        return '200-300'
    elif fare < 400:
        return '300-400'
    else:
        return '400-500'
    
df2['fare category'] = df2.fare.apply(func=fare_classification)    

print(df2.to_string())
```

```python
# How to create Pivot Table in Pandas

df2.pivot_table(index=['sex','class','embark_town'],aggfunc={'age':'mean', 'class':'count', 'fare':'sum'})
```
